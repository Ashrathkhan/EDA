# **Project Summary -**

The primary objective of this Exploratory Data Analysis (EDA) project is to explore, clean, and analyze the Amazon Prime Video dataset to gain insights into the platform’s content catalog. This analysis involves examining various attributes such as content type (Movie or TV Show), genres, release years, age certifications, user ratings, runtime, and popularity metrics. By performing this analysis, we aim to understand the trends and characteristics of the content hosted on Amazon Prime Video and draw meaningful conclusions that could assist content curators, marketers, or platform developers.
Dataset Overview
The project uses two datasets: titles.csv, which contains metadata about each title including name, type, release year, rating, genres, etc., and credits.csv, which contains casting and crew details. These datasets are merged using a common identifier (id) to create a comprehensive data frame for analysis.

**Data cleaning**
Initial steps involve checking for missing values and duplicate entries. Columns like description, age_certification, imdb_id, imdb_score, and tmdb_score contained missing values. Different imputation strategies were applied:

Mode imputation for categorical columns like age_certification.

Mean imputation for numerical columns like imdb_score and tmdb_popularity.

Dropping rows where essential data like description is missing.

Duplicate rows were also removed to maintain data integrity. The data was then verified for consistency, and irrelevant or unnecessary columns were ignored during visualization.

 **Univariate Analysis**
Univariate visualizations were performed to understand the distribution of each variable individually:

**Content Type**: Count plot showed that movies dominate the catalog over TV shows.

**Age Certification:** Most content falls under "PG", "TV-14", and "R", indicating a wide variety in audience targeting.

**Genres:** Dramas, comedies, and documentaries are among the most common genres.

**Release Year:** Content production increased significantly after 2010, peaking around 2020.

**Runtime:** Movies typically range between 60 to 120 minutes; TV shows are variable depending on episodes.

**IMDb & TMDB Scores:** Most content scores between 6 to 8, indicating moderate to good user reception.

 **key insights**
**Dominance of Movie Content:** Amazon Prime hosts a larger number of movies than TV series.

**Age Certification Trends:** A significant portion of content is aimed at teenagers and adults, with relatively fewer titles for younger audiences.

**Genre Distribution:** Drama and comedy dominate the platform, followed by thriller and documentary content.

**Popularity vs Ratings:** Not all highly rated content is popular and vice versa, indicating that popularity may depend on promotion, recency, or cast rather than just quality.

**Content Growth:** A rise in content production is seen post-2015, aligning with the global boom in streaming services.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
merged_df.dropna(subset=['description'], inplace=True)
````

````python
merged_df['age_certification'].mode()
````

````python
merged_df['age_certification'].fillna(merged_df['age_certification'].mode()[0], inplace=True)
````

````python
merged_df['seasons'].fillna(0, inplace=True)
````

````python
round(merged_df['imdb_score'].mean(),1)
merged_df['imdb_score'].fillna(round(merged_df['imdb_score'].mean(),1),inplace=True)
````

````python
round(merged_df['tmdb_score'].mean(),1)
merged_df['tmdb_score'].fillna(round(merged_df['tmdb_score'].mean(),1),inplace=True)
````

````python
round(merged_df['tmdb_popularity'].mean(),2)
````

````python
merged_df['tmdb_popularity'].fillna(round(merged_df['tmdb_popularity'].mean(),2),inplace=True)
````

````python
merged_df.dropna(subset=['imdb_id'], inplace=True)
````

````python
merged_df.dropna(subset=['character'], inplace=True)
````

````python
merged_df.isnull().sum()
````

````python
merged_df.drop_duplicates(inplace=True)
merged_df.duplicated().sum()
````

#### Chart - 1

````python
# Chart - 1 visualization code
plt.figure(figsize=(8,5))
sns.boxplot(data=merged_df, x='imdb_score', color='skyblue')
plt.title('Box Plot of IMDB Score')
plt.show()
````

##### 1. Why did you pick the specific chart?

A box plot is ideal for visualizing the distribution, central tendency, and spread of numerical data like imdb_score.

**It quickly highlights:**
Median (central value)
Interquartile range (IQR) (middle 50% of scores)
Outliers (very low or very high ratings).

##### 2. What is/are the insight(s) found from the chart?

1.Median IMDB Score is around 6, showing most titles are rated moderately.

2.IQR (box) ranges from approx 5 to 7, indicating a tight concentration of scores.

3.Outliers are clearly visible below 4 and above 8.5, suggesting a few titles are rated poorly or exceptionally well.

4.Symmetry of the box indicates that the data is fairly normally distributed, but with some left-side skew due to more low outliers.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**Positive Business Growth Indicators:**
Strong central tendency (6+ rating) shows that most titles are well received.

Small IQR means consistent quality, which boosts platform credibility and customer satisfaction.

High-rating outliers (8.5+) show some content has viral or blockbuster potential.

**Negative Growth Indicators:**
The presence of many low-rating outliers (<4) may hurt reputation if these are recent or heavily promoted titles.

If poor content is not removed or improved, it may lead to user churn.
